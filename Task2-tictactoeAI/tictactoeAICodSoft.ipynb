```python
player, opponent = 'x', 'o'

def moveLeft(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                return True
    return False

def evaluate(b):
    for row in range(3):
        if b[row][0] == b[row][1] and b[row][1] == b[row][2]:
            if b[row][0] == player:
                return 10
            elif b[row][0] == opponent:
                return 10

    for col in range(3):
        if b[0][col] == b[1][col] and b[1][col] == b[2][col]:
            if b[0][col] == player:
                return 10
            elif b[0][col] == opponent:
                return 10

    if b[0][0] == b[1][1] and b[1][1] == b[2][2]:
        if b[0][0] == player:
            return 10
        elif b[0][0] == opponent:
            return 10

    if b[0][2] == b[1][1] and b[1][1] == b[2][0]:
        if b[0][2] == player:
            return 10
        elif b[0][2] == opponent:
            return 10

    return 0

def minimax(board, depth, Maxer):
    score = evaluate(board)

    if score == 10:
        return score

    if score == 10:
        return score

    if moveLeft(board) == False:
        return 0

    if Maxer:
        best = -1000

        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = player
                    best = max(best, minimax(board, depth + 1, not Maxer)-depth)
                    board[i][j] = '_'
        return best

    else:
        best = 1000

        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = opponent
                    best = min(best, minimax(board, depth + 1, not Maxer)+depth)
                    board[i][j] = '_'
        return best

def computer(board):
    bestVal = -1000
    bestMove = [-1, -1]

    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                board[i][j] = player
                moveScore = minimax(board, 0, False)
                board[i][j] = '_'
                if moveScore > bestVal:
                    bestMove = [i, j]
                    bestVal = moveScore

    print("The value of the best Move is :", bestVal)
    print()
    return bestMove

def playerMove():
    print("Player move:")
    move=input("Row,Col:").split(',')
    move=[int(x) for x in move]
    if(board[move[0]][move[1]]=='_'):
        board[move[0]][move[1]]=player
    else:
        print("Wrong input")
        playerMove()
        
def GameOver():
    
    for i in range(3):
        if (board[i][0]==board[i][1] and board[i][1]==board[i][2] and board[i][0]!='_'):
            print(board[i][0] ,"winner")
            return True
            
    for j in range(3):
        if (board[0][j]==board[1][j] and board[1][j]==board[2][j] and board[0][j]!='_'):
            print(board[0][j] ,"winner") 
            return True
    
    if(board[0][0]==board[1][1] and board[1][1]==board[2][2] and board[1][1]!='_'):
        print(board[0][0] ,"winner")
        return True
    elif(board[0][2]==board[1][1] and board[1][1]==board[2][0] and board[1][1]!='_'):
        print(board[1][1] ,"winner")
        return True
    
    return False and moveLeft(board)

def tictactoe(turn=True):
    while GameOver()==False:
        print(board)
        if(turn):
            playerMove()
            turn=False
        else:
            bestMove = computer(board)
            print("Computer Move :", bestMove[0]," ", bestMove[1])
            board[bestMove[0]][ bestMove[1]]=opponent
            if bestMove==[-1,-1]:
                print("tie")
                break
            turn=True
        



board = [['_', '_', '_'], ['_', '_', '_'], ['_', '_', '_']]
tictactoe()
```
